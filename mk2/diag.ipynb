```python
import sqlite3


planes = sqlite3.connect("planes.sqb")
```

```python
def get_data_for_airframe(airframe):
    choose = f"SELECT ts,tail from planes where hex='{airframe}'"
    print(choose)
    cur = planes.cursor()
    cur.execute(choose)
    rows = cur.fetchall()
    prev = 0
    for record in rows:
        if prev == 0:
            prev = record[0]
        else:
            interval = record[0]-prev
            if interval < 3000:
                print(f" diff {int((record[0]-prev)/60)} {record}")
            prev = record[0]
        
```

```python
hex_codes = planes.execute("SELECT DISTINCT hex from planes")

for airframe in hex_codes:
    print(airframe[0])
    get_data_for_airframe(airframe[0])
    
    
```
